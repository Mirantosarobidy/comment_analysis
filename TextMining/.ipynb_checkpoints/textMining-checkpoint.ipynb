```python
#remove the URls and make all the things lowercase
def URL_clean(data):
    return re.sub('((www.[^s]+)|(https?://[^s]+))',' ',data.lower())
```

```python
#remove the number and make all the things lowercase
def number_clean(data):
    return re.sub('[0-9]+', '', data.lower())
```

```python
#remove the ponctuation
def ponctuation_clean(data):
    return re.sub('[^\w\s]', '', data.lower())
```
