# Cleaning dataset

```python
caractere_inutil = set()
for i in range(len(data['clean_text'])):
    caractere = str(data['clean_text'][i]).translate(str(data['clean_text'][i]).maketrans('','',string.punctuation+"0123456789azertyuiopqsdfghjklmwxcvbn")).lower()
    caractere = caractere.split(' ')
    caractere_inutil.add(''.join(caractere))
```

```python
inutil = ''
for c in caractere_inutil:
    inutil += c
```

```python
inutil
```

```python
data_clean = []
for i in range(len(data['clean_text'])):
    clean = str(data['clean_text'][i]).translate(str(data['clean_text'][i]).maketrans('','',string.punctuation+inutil+'0123456789')).lower()
    words = word_tokenize(clean)
    #stopwords
    clean_stopword =' '.join([word for word in words if not word in stopwords.words('english')])
    data_clean.append(clean_stopword)
```

```python
data_text = pd.DataFrame(data_clean, columns = ['Text'])
data_text
```

```python
cleaning_data = data_text.merge(data, how='inner', left_index=True, right_index=True)
cleaning_data = cleaning_data.drop(['clean_text'], axis=1)
cleaning_data
```

```python
cleaning_data.to_csv('data.csv', index=False)
```
